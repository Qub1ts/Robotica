```python
import sys
sys.path.append('./python_code')
import select_pixels as sel
```

```python
import cv2
from imageio import imread, imsave
from matplotlib import pyplot as plt
```

```python
# Captura de imagenes desde la camara del ordenador
capture = cv2.VideoCapture('video2017-3.avi')
```

```python
key = 0
while (key != 27):
    # Leo una imagen del video
    ret,im = capture.read()
 
    if not ret:
        print('Se acabo el video.')
        cv2.destroyAllWindows()
        break
        
    # Muestro la imagen en una ventana OpenCV
    cv2.imshow("Captura", im)

    key = cv2.waitKey(35)
print('Key:',key)

# Cierro la ventana 
cv2.destroyAllWindows() 
capture.release()
```

```python
# Transformo la imagen al formato Python
imNumpy=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

# Llamo a la funcion para etiquetar la imagen
markImg = sel.select_fg_bg(imNumpy)

plt.imshow(markImg)
plt.show()
```

```python
# Guardo las imagenes
imsave('imagen_original.png', imNumpy)
imsave('imagen_marcada.png', markImg)
```
